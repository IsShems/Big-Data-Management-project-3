## Setup

Install helpers and define service addresses.

````python
import subprocess, sys

def _ensure(pkg, import_name=None):
    import importlib.util
    if importlib.util.find_spec(import_name or pkg) is None:
        subprocess.check_call([sys.executable, '-m', 'pip', 'install', '-q', pkg])

_ensure('requests')
_ensure('psycopg2-binary', 'psycopg2')
print('packages ready')
````

````python
import os, json, time, requests
import psycopg2

# ── Service endpoints ────────────────────────────────────────────────────────
AIRFLOW_URL  = 'http://airflow:8080'
CONNECT_URL  = 'http://connect:8083'
BOOTSTRAP    = 'kafka:9092'
DAG_ID       = 'cdc_pipeline'
CONNECTOR_ID = os.environ.get('CDC_CONNECTOR_NAME', 'pg-cdc-connector')
AUTH         = (os.environ.get('AIRFLOW_USER', 'admin'),
                os.environ.get('AIRFLOW_PASSWORD', 'admin'))
DAGS_DIR     = '/home/jovyan/project/dags'

# ── PostgreSQL helper ────────────────────────────────────────────────────────
PG = dict(host='postgres', port=5432, dbname='sourcedb',
          user=os.environ.get('PG_USER', 'cdc_user'),
          password=os.environ.get('PG_PASSWORD', 'admin'))

def pg(sql, params=None, fetch=False):
    conn = psycopg2.connect(**PG)
    conn.autocommit = True
    cur = conn.cursor()
    cur.execute(sql, params)
    result = cur.fetchall() if fetch else None
    cur.close(); conn.close()
    return result

# ── Airflow REST API helpers ─────────────────────────────────────────────────
def af_get(path):
    return requests.get(f'{AIRFLOW_URL}{path}', auth=AUTH)

def af_post(path, payload):
    return requests.post(f'{AIRFLOW_URL}{path}', json=payload, auth=AUTH)

def af_patch(path, payload):
    return requests.patch(f'{AIRFLOW_URL}{path}', json=payload, auth=AUTH)

def trigger_dag(dag_id=DAG_ID, logical_date=None):
    payload = {'dag_run_id': f'notebook__{int(time.time())}'}
    if logical_date:
        payload['logical_date'] = logical_date
    r = af_post(f'/api/v1/dags/{dag_id}/dagRuns', payload)
    data = r.json()
    if r.status_code == 200:
        print(f'Triggered  run_id={data["dag_run_id"]}  state={data["state"]}')
        return data['dag_run_id']
    print(f'Error {r.status_code}: {r.text[:300]}')
    return None

def poll_run(run_id, dag_id=DAG_ID, timeout=300, interval=5):
    print(f'Polling {dag_id} / {run_id} ...')
    for elapsed in range(0, timeout, interval):
        r = af_get(f'/api/v1/dags/{dag_id}/dagRuns/{run_id}')
        state = r.json().get('state', 'unknown')
        print(f'  [{elapsed:>3}s] {state}')
        if state in ('success', 'failed'):
            return state
        time.sleep(interval)
    print('  TIMEOUT')
    return 'timeout'

def show_tasks(run_id, dag_id=DAG_ID):
    r = af_get(f'/api/v1/dags/{dag_id}/dagRuns/{run_id}/taskInstances')
    tasks = r.json().get('task_instances', [])
    icons = {'success': '\u2713', 'failed': '\u2717', 'skipped': '\u23ed',
             'running': '\u27f3', 'upstream_failed': '!'}
    print(f'Task states  {run_id[:50]}:')
    for t in sorted(tasks, key=lambda x: x.get('start_date') or ''):
        s = t.get('state') or 'none'
        print(f'  {icons.get(s, "?")} {t["task_id"]:<28} {s}')

print('Helpers ready.')
````

````python
# Verify all services before starting

r = requests.get(f'{AIRFLOW_URL}/api/v1/health', auth=AUTH)
h = r.json()
print(f'Airflow    scheduler={h.get("scheduler",{}).get("status")}  '
      f'db={h.get("metadatabase",{}).get("status")}')

r = requests.get(f'{CONNECT_URL}/')
print(f'Connect    version={r.json().get("version")}')

ver = pg('SELECT version();', fetch=True)
print(f'Postgres   {ver[0][0][:55]}...')

wal = pg('SHOW wal_level;', fetch=True)
assert wal[0][0] == 'logical', 'wal_level must be logical for CDC!'
print(f'wal_level  {wal[0][0]}  \u2713')
print('\nAll services healthy.')
````

## Part 1 - HttpSensor: Health-Checking the Debezium Connector

**Why a sensor before any data task?**

CDC events only flow when the Debezium connector is `RUNNING`. If we start reading
from Kafka before the connector is healthy we read zero events and write an empty Bronze
file  a silent data loss. An `HttpSensor` guards against this by polling the
Kafka Connect REST API until the connector reports `RUNNING`.

```python
health_check = HttpSensor(
    task_id        = 'health_check',
    http_conn_id   = 'debezium_connect',   # Airflow Connection object
    endpoint       = '/connectors/pg-cdc-connector/status',
    response_check = lambda r: r.json()['connector']['state'] == 'RUNNING',
    poke_interval  = 20,   # seconds between polls
    timeout        = 300,  # fail with AirflowSensorTimeout after 5 min
    mode           = 'poke',
)
```

**`http_conn_id`** references a *Connection* stored in Airflow's metadata DB.
It bundles `host`, `port`, and `schema` so the DAG code stays environment-agnostic.
The next cell creates it via the Airflow REST API.

````python
# 1. Ensure the Debezium connector is registered and running
status = requests.get(f'{CONNECT_URL}/connectors/{CONNECTOR_ID}/status')
if status.status_code != 200:
    print('Connector not found  registering...')
    cfg = {
        'name': CONNECTOR_ID,
        'config': {
            'connector.class': 'io.debezium.connector.postgresql.PostgresConnector',
            'database.hostname': 'postgres',
            'database.port': '5432',
            'database.user': PG['user'],
            'database.password': PG['password'],
            'database.dbname': 'sourcedb',
            'topic.prefix': 'dbserver1',
            'table.include.list': 'public.customers',
            'plugin.name': 'pgoutput',
            'snapshot.mode': 'initial',
            'key.converter.schemas.enable': 'false',
            'value.converter.schemas.enable': 'false',
        }
    }
    r = requests.post(f'{CONNECT_URL}/connectors',
                      headers={'Content-Type': 'application/json'},
                      data=json.dumps(cfg))
    print(f'  created: {r.status_code}')
    time.sleep(8)
else:
    print(f'Connector state: {status.json()["connector"]["state"]}')

# 2. Create the Airflow Connection that HttpSensor will use
requests.delete(f'{AIRFLOW_URL}/api/v1/connections/debezium_connect', auth=AUTH)
r = af_post('/api/v1/connections', {
    'connection_id': 'debezium_connect',
    'conn_type': 'http',
    'host': 'connect',
    'port': 8083,
    'schema': 'http',
})
print(f'Connection \'debezium_connect\' created: HTTP {r.status_code}')
````

````python
# Manually simulate the response_check the sensor will run
url = f'{CONNECT_URL}/connectors/{CONNECTOR_ID}/status'
r = requests.get(url)
print(f'GET {url}')
print(f'HTTP {r.status_code}')
print(json.dumps(r.json(), indent=2))

passes = r.json().get('connector', {}).get('state') == 'RUNNING'
_ok = '\u2713'
_retry = '\u21ba'
print(f'\nresponse_check -> {passes}  '
      f'(sensor would {"succeed " + _ok if passes else "retry " + _retry})')
````

## Part 2 - DAG Structure with Dependency Chains

The full pipeline DAG looks like this:

```
health_check
     |
check_new_events  (BranchPythonOperator)
    /          \
bronze_cdc    skip_merge
    |
silver_cdc
```

**Dependency rules:**
- `health_check` must pass before any data work starts
- `check_new_events` decides at runtime which branch to take
- `bronze_cdc` reads new Kafka events and writes a JSON checkpoint (keyed by logical date)
- `silver_cdc` reads that checkpoint and applies UPSERT/DELETE to PostgreSQL
- `skip_merge` is an `EmptyOperator` - a no-op placeholder so the run still succeeds cleanly

**Key properties:**
- `catchup=False` - no automatic backfill on deploy
- `max_active_runs=1` - prevents concurrent runs that could corrupt the offset state
- `schedule_interval='@hourly'` - triggers once per hour

The next cell writes this DAG to `dags/` (shared mount with the Airflow container).

````python
dag_code = '''
from datetime import datetime, timedelta
import os, json, subprocess, sys

from airflow import DAG
from airflow.operators.python import PythonOperator, BranchPythonOperator
from airflow.operators.empty import EmptyOperator
from airflow.providers.http.sensors.http import HttpSensor

BOOTSTRAP  = 'kafka:9092'
TOPIC      = 'dbserver1.public.customers'
CONNECTOR  = 'pg-cdc-connector'
STATE_FILE = '/opt/airflow/dags/cdc_state.json'


def _pip(pkg, import_as=None):
    import importlib.util
    if importlib.util.find_spec(import_as or pkg) is None:
        subprocess.check_call([sys.executable, '-m', 'pip', 'install', '-q', pkg])


# ── BranchPythonOperator ─────────────────────────────────────────────────────

def check_new_events(**ctx):
    """Compare latest Kafka offset to stored offset; return next task_id."""
    _pip('kafka-python-ng', 'kafka')
    from kafka import KafkaConsumer, TopicPartition

    consumer = KafkaConsumer(bootstrap_servers=BOOTSTRAP)
    parts = consumer.partitions_for_topic(TOPIC) or set()
    if not parts:
        consumer.close()
        ctx['ti'].xcom_push(key='new_count', value=0)
        return 'skip_merge'

    tp = TopicPartition(TOPIC, 0)
    consumer.assign([tp])
    consumer.seek_to_end(tp)
    latest = consumer.position(tp)
    consumer.close()

    state = {}
    if os.path.exists(STATE_FILE):
        with open(STATE_FILE) as f:
            state = json.load(f)

    stored    = state.get('last_offset', 0)
    new_count = max(0, latest - stored)
    print('Kafka offsets: latest=%d  stored=%d  new=%d' % (latest, stored, new_count))

    ctx['ti'].xcom_push(key='from_offset', value=stored)
    ctx['ti'].xcom_push(key='to_offset',   value=latest)
    ctx['ti'].xcom_push(key='new_count',   value=new_count)
    return 'bronze_cdc' if new_count > 0 else 'skip_merge'


# ── Bronze CDC ───────────────────────────────────────────────────────────────

def run_bronze(**ctx):
    """Read new Kafka events; write Bronze JSON checkpoint (idempotent by logical date)."""
    _pip('kafka-python-ng', 'kafka')
    from kafka import KafkaConsumer, TopicPartition

    ti          = ctx['ti']
    from_offset = ti.xcom_pull(key='from_offset', task_ids='check_new_events')
    to_offset   = ti.xcom_pull(key='to_offset',   task_ids='check_new_events')

    if from_offset is None or from_offset >= to_offset:
        print('Nothing to read')
        return 0

    consumer = KafkaConsumer(
        bootstrap_servers=BOOTSTRAP,
        auto_offset_reset='earliest',
        enable_auto_commit=False,
    )
    tp = TopicPartition(TOPIC, 0)
    consumer.assign([tp])
    consumer.seek(tp, from_offset)

    events = []
    while consumer.position(tp) < to_offset:
        batch = consumer.poll(timeout_ms=3000, max_records=500)
        if not batch:
            break
        for msgs in batch.values():
            for msg in msgs:
                if msg.offset < to_offset and msg.value:
                    events.append({'offset': msg.offset, 'value': msg.value.decode()})
    consumer.close()

    # Key the file by logical date -> re-running the same logical date overwrites safely
    ds          = ctx['ds_nodash']
    bronze_path = '/opt/airflow/dags/bronze_' + ds + '.json'
    with open(bronze_path, 'w') as f:
        json.dump({'from': from_offset, 'to': to_offset, 'events': events}, f)

    with open(STATE_FILE, 'w') as f:
        json.dump({'last_offset': to_offset, 'last_run': ctx['ds']}, f)

    print('Bronze: %d events -> %s' % (len(events), bronze_path))
    return len(events)


# ── Silver CDC ───────────────────────────────────────────────────────────────

def run_silver(**ctx):
    """Apply Bronze events to silver_customers_af via PostgreSQL ON CONFLICT upsert."""
    _pip('psycopg2-binary', 'psycopg2')
    import psycopg2

    ds          = ctx['ds_nodash']
    bronze_path = '/opt/airflow/dags/bronze_' + ds + '.json'
    if not os.path.exists(bronze_path):
        print('No bronze file for ' + ds)
        return 0

    with open(bronze_path) as f:
        data = json.load(f)

    conn = psycopg2.connect(
        host='postgres', port=5432, dbname='sourcedb',
        user=os.environ.get('PG_USER', 'cdc_user'),
        password=os.environ.get('PG_PASSWORD', 'admin'),
    )
    conn.autocommit = True
    cur = conn.cursor()

    create_ddl = (
        'CREATE TABLE IF NOT EXISTS silver_customers_af ('
        'id INTEGER PRIMARY KEY, '
        'name VARCHAR(100), '
        'email VARCHAR(200), '
        'country VARCHAR(50), '
        'updated_at TIMESTAMP DEFAULT NOW())'
    )
    cur.execute(create_ddl)

    upsert_sql = (
        'INSERT INTO silver_customers_af (id, name, email, country, updated_at) '
        'VALUES (%s, %s, %s, %s, NOW()) '
        'ON CONFLICT (id) DO UPDATE SET '
        'name=EXCLUDED.name, email=EXCLUDED.email, '
        'country=EXCLUDED.country, updated_at=EXCLUDED.updated_at'
    )

    applied = 0
    for evt in data.get('events', []):
        try:
            payload = json.loads(evt['value']).get('payload', {})
            op     = payload.get('op')
            after  = payload.get('after')  or {}
            before = payload.get('before') or {}

            if op in ('c', 'r', 'u') and after.get('id'):
                cur.execute(upsert_sql,
                            (after['id'], after.get('name'),
                             after.get('email'), after.get('country')))
                applied += 1
            elif op == 'd' and before.get('id'):
                cur.execute('DELETE FROM silver_customers_af WHERE id=%s', (before['id'],))
                applied += 1
        except Exception as e:
            print('  skip offset=%s: %s' % (evt.get('offset'), e))

    cur.close()
    conn.close()
    print('Silver: %d operations applied' % applied)
    return applied


# ── DAG definition ────────────────────────────────────────────────────────────

default_args = {
    'retries': 3,
    'retry_delay': timedelta(minutes=1),
    'retry_exponential_backoff': True,
    'max_retry_delay': timedelta(minutes=10),
    'sla': timedelta(minutes=30),
}

with DAG(
    dag_id='cdc_pipeline',
    description='CDC: PostgreSQL -> Debezium -> Kafka -> Bronze -> Silver',
    schedule_interval='@hourly',
    start_date=datetime(2026, 4, 19, 0, 0, 0),
    catchup=False,
    max_active_runs=1,
    default_args=default_args,
    tags=['cdc', 'debezium', 'week06'],
) as dag:

    health_check = HttpSensor(
        task_id='health_check',
        http_conn_id='debezium_connect',
        endpoint='/connectors/' + CONNECTOR + '/status',
        response_check=lambda r: r.json().get('connector', {}).get('state') == 'RUNNING',
        poke_interval=20,
        timeout=300,
        mode='poke',
    )

    branch = BranchPythonOperator(
        task_id='check_new_events',
        python_callable=check_new_events,
    )

    bronze = PythonOperator(
        task_id='bronze_cdc',
        python_callable=run_bronze,
    )

    silver = PythonOperator(
        task_id='silver_cdc',
        python_callable=run_silver,
    )

    skip = EmptyOperator(task_id='skip_merge')

    health_check >> branch >> [bronze, skip]
    bronze >> silver
'''.lstrip()

dag_path = DAGS_DIR + '/cdc_pipeline.py'
with open(dag_path, 'w') as f:
    f.write(dag_code)
print(f'DAG written to {dag_path}')
print('Airflow scans the dags/ folder every ~30 seconds  wait before the next cell.')
````

````python
# Wait for Airflow to pick up the DAG, then unpause it
print('Waiting 35 s for file scan ...')
time.sleep(35)

# Unpause (new DAGs start paused)
r = af_patch(f'/api/v1/dags/{DAG_ID}', {'is_paused': False})
print(f'Unpause HTTP {r.status_code}: paused={r.json().get("is_paused")}')

# Show the DAG summary from Airflow
r = af_get(f'/api/v1/dags/{DAG_ID}')
d = r.json()
print(f'\nDAG: {d.get("dag_id")}')
print(f'  schedule_interval : {d.get("schedule_interval")}')
print(f'  is_paused         : {d.get("is_paused")}')
print(f'  max_active_runs   : {d.get("max_active_runs")}')
print(f'  tags              : {[t["name"] for t in d.get("tags", [])]}')
````

````python
# Show the task dependency graph returned by the API
r = af_get(f'/api/v1/dags/{DAG_ID}/tasks')
tasks = r.json().get('tasks', [])

print(f'Tasks in {DAG_ID}:')
for t in tasks:
    deps = t.get('downstream_task_ids', [])
    arrow = ' -> ' + str(deps) if deps else ''
    class_ref = t.get('class_ref', {}).get('class_name', t.get('task_type', '?'))
    print(f'  {t["task_id"]:<28} [{class_ref}]{arrow}')
````

## Part 3 - Scheduling, Retries & SLA

The DAG is configured with production-grade reliability settings:

| Parameter | Value | What it does |
|---|---|---|
| `schedule_interval` | `@hourly` | Runs once per hour; logical date = start of the hour |
| `catchup=False` |  | Does **not** run missed intervals on deploy |
| `max_active_runs=1` |  | No parallel runs; prevents offset state corruption |
| `retries=3` |  | Failed task is retried 3 times before the run fails |
| `retry_delay` | 1 min | Wait before first retry |
| `retry_exponential_backoff=True` |  | Delay doubles each retry (1 min, 2 min, 4 min) |
| `max_retry_delay` | 10 min | Caps the backoff so retries don\u2019t wait indefinitely |
| `sla` | 30 min | Airflow fires a callback if the task runs longer than 30 minutes |

**Idempotency** is built into both Bronze and Silver:
- Bronze writes to `bronze_{ds_nodash}.json`  re-running the same logical date overwrites the file
- Silver uses `ON CONFLICT (id) DO UPDATE`  applying the same events twice is a no-op
- The stored offset (`cdc_state.json`) only advances after a successful Bronze write

Let\u2019s trigger the first manual run and watch it succeed.

````python
# Create the customers table if it doesn't exist, then seed if empty
pg('''
    CREATE TABLE IF NOT EXISTS customers (
        id         SERIAL PRIMARY KEY,
        name       VARCHAR(100) NOT NULL,
        email      VARCHAR(200) NOT NULL,
        country    VARCHAR(50)  NOT NULL,
        created_at TIMESTAMP DEFAULT NOW()
    );
''')

existing = pg('SELECT COUNT(*) FROM customers;', fetch=True)
if existing[0][0] == 0:
    pg("""INSERT INTO customers (name, email, country) VALUES
        ('Alice Mets',   'alice@example.com',  'Estonia'),
        ('Bob Virtanen', 'bob@example.com',    'Finland'),
        ('Carol Ozols',  'carol@example.com',  'Latvia')""")
    print('Seeded 3 customers')
    time.sleep(5)
else:
    print(f'customers table already has {existing[0][0]} rows')

# Trigger the first run
run1 = trigger_dag()
assert run1, 'Trigger failed  is the DAG visible in Airflow? Check the UI.'
````

````python
# Poll until the run reaches a terminal state (success or failed)
# Expect: health_check -> check_new_events -> bronze_cdc -> silver_cdc (all success)
# First run installs kafka-python-ng and psycopg2 inside the Airflow container (~30-60s extra)
final = poll_run(run1, timeout=360)
print(f'\nFinal state: {final}')
show_tasks(run1)
````

````python
# Inspect the Silver table in PostgreSQL
rows = pg('SELECT id, name, email, country FROM silver_customers_af ORDER BY id;', fetch=True)
print(f'silver_customers_af ({len(rows)} rows):')
for r in rows:
    print(f'  {r}')

# Read the state file (written by the Bronze task inside the Airflow container)
state_path = DAGS_DIR + '/cdc_state.json'
if os.path.exists(state_path):
    with open(state_path) as f:
        state = json.load(f)
    print(f'\ncdc_state.json: {state}')
else:
    print('\ncdc_state.json not found yet  did the run succeed?')
````

## Part 4 - Backfill & Idempotency

**Backfill scenario:** the pipeline was down for a period and we need to reprocess
data from scratch. In Airflow you can run `airflow dags backfill -s <start> -e <end>`,
but for this exercise we simulate it by resetting the stored offset to 0  forcing
the next run to read all Kafka events from the beginning.

**Idempotency guarantee:** re-processing the same events must produce the same Silver
table. This holds because:
1. Bronze overwrites the same JSON file (keyed by `ds_nodash`)
2. Silver uses `ON CONFLICT (id) DO UPDATE`  an upsert is safe to apply N times

````python
# Make some changes in PostgreSQL so there are new events to capture
pg("INSERT INTO customers (name, email, country) VALUES ('David Jonaitis', 'david@example.com', 'Lithuania');")
pg("UPDATE customers SET email='alice.updated@example.com' WHERE name='Alice Mets';")
pg("DELETE FROM customers WHERE name='Carol Ozols';")

rows = pg('SELECT id, name, email, country FROM customers ORDER BY id;', fetch=True)
print('PostgreSQL customers (after changes):')
for r in rows:
    print(f'  {r}')
print('\nWaiting 5s for Debezium ...')
time.sleep(5)
````

````python
# Trigger run 2 — picks up the new events
run2 = trigger_dag()
final = poll_run(run2, timeout=240)
print(f'\nFinal state: {final}')
show_tasks(run2)

print('\nSilver after run 2:')
rows = pg('SELECT id, name, email, country FROM silver_customers_af ORDER BY id;', fetch=True)
for r in rows:
    print(f'  {r}')
````

````python
# ── Backfill simulation ──────────────────────────────────────────────────────
# Reset the stored offset to 0 to force full reprocessing
state_path = DAGS_DIR + '/cdc_state.json'
with open(state_path, 'w') as f:
    json.dump({'last_offset': 0, 'last_run': 'reset_for_backfill'}, f)
print('State reset: last_offset = 0 (simulating a backfill)')

# Also remove the existing bronze checkpoint so Silver re-applies from scratch
import glob
for bf in glob.glob(DAGS_DIR + '/bronze_*.json'):
    os.remove(bf)
    print(f'Removed {bf}')

# Clear Silver so we can see the reprocessing rebuild it
pg('DELETE FROM silver_customers_af WHERE 1=1;')
print('Silver cleared.')
````

````python
# Trigger run 3 — reads ALL events from offset 0, rebuilds Silver
run3 = trigger_dag()
final = poll_run(run3, timeout=240)
print(f'\nFinal state: {final}')
show_tasks(run3)

# The Silver table should match PostgreSQL exactly
print('\nSilver after backfill run:')
silver_rows = pg('SELECT id, name, email, country FROM silver_customers_af ORDER BY id;', fetch=True)
for r in silver_rows:
    print(f'  {r}')

pg_rows = pg('SELECT id, name, email, country FROM customers ORDER BY id;', fetch=True)
print(f'\nPostgreSQL: {len(pg_rows)} rows | Silver: {len(silver_rows)} rows')
match = set(silver_rows) == set(pg_rows)
_yes = '\u2713 YES'
_no = '\u2717 NO  check the logs'
print(f'Match: {_yes if match else _no}')
````

## Part 5 - Failure Test: Connector Down

What happens if Debezium is unavailable when a run starts?

**Expected sequence:**
1. `health_check` sensor polls the Connect API every 20 seconds
2. Each poll returns 404 (connector deleted)  sensor stays in `running` state
3. After `timeout=300s`, sensor raises `AirflowSensorTimeout`  run fails
4. With `retries=3`, Airflow will retry the **task** (not the run)  sensor polls again
5. When we re-register the connector, the sensor succeeds on the next poke

We\u2019ll delete the connector, trigger a run, watch it struggle, then re-register to let it recover.

````python
# Delete the connector to simulate a failure
r = requests.delete(f'{CONNECT_URL}/connectors/{CONNECTOR_ID}')
print(f'DELETE connector: HTTP {r.status_code}')

# Confirm it's gone
time.sleep(2)
r = requests.get(f'{CONNECT_URL}/connectors/{CONNECTOR_ID}/status')
print(f'Status check: HTTP {r.status_code}  (404 = connector removed)')

# Trigger a DAG run with the connector down
run_fail = trigger_dag()
print(f'\nRun triggered: {run_fail}')
print('The health_check sensor will now retry every 20s...')
````

````python
# Observe the sensor retrying for ~60 seconds
print('Watching health_check retry for 60s ...')
for i in range(12):
    r = af_get(f'/api/v1/dags/{DAG_ID}/dagRuns/{run_fail}')
    run_state = r.json().get('state', '?')

    r2 = af_get(f'/api/v1/dags/{DAG_ID}/dagRuns/{run_fail}/taskInstances')
    tasks = {t['task_id']: t.get('state', 'none')
             for t in r2.json().get('task_instances', [])}
    sensor_state = tasks.get('health_check', 'not_started')

    print(f'  [{i*5:>3}s] run={run_state}  health_check={sensor_state}')
    if run_state in ('failed', 'success'):
        break
    time.sleep(5)

print('\nNow re-registering the connector ...')
````

````python
# Re-register the connector
cfg = {
    'name': CONNECTOR_ID,
    'config': {
        'connector.class': 'io.debezium.connector.postgresql.PostgresConnector',
        'database.hostname': 'postgres',
        'database.port': '5432',
        'database.user': PG['user'],
        'database.password': PG['password'],
        'database.dbname': 'sourcedb',
        'topic.prefix': 'dbserver1',
        'table.include.list': 'public.customers',
        'plugin.name': 'pgoutput',
        'snapshot.mode': 'never',   # don't re-snapshot; continue from WAL
        'key.converter.schemas.enable': 'false',
        'value.converter.schemas.enable': 'false',
    }
}
r = requests.post(f'{CONNECT_URL}/connectors',
                  headers={'Content-Type': 'application/json'},
                  data=json.dumps(cfg))
print(f'Re-register connector: HTTP {r.status_code}')
time.sleep(8)

r = requests.get(f'{CONNECT_URL}/connectors/{CONNECTOR_ID}/status')
print(f'Connector state: {r.json()["connector"]["state"]}')
print('\nThe next sensor poke (within 20s) should succeed.')
print('Trigger a new run so we can observe the recovery cleanly.')
````

````python
# Trigger a recovery run — should succeed end-to-end
run_recover = trigger_dag()
final = poll_run(run_recover, timeout=300)
print(f'\nFinal state: {final}')
show_tasks(run_recover)
````

## Part 6 - BranchPythonOperator: Skip MERGE When Idle

In production, most hourly runs will find **no new events** in Kafka (if data is bursty or
the source is quiet). Running an expensive MERGE for zero rows wastes resources.

The `BranchPythonOperator` lets the task function decide at runtime which downstream
task to activate. Tasks **not** chosen are automatically marked `skipped`.

```
check_new_events()          returns 'bronze_cdc'   → bronze + silver run
                            returns 'skip_merge'   → EmptyOperator runs, silver skipped
```

**Key invariant:** the function must return a valid `task_id` (string) or a list of
task IDs. Returning an unknown ID raises `AirflowException`.

After the backfill in Part 4, the stored offset equals the latest Kafka offset 
so the next run should take the `skip_merge` path without any code changes.

````python
# Confirm: no new changes in PostgreSQL since the last run
state_path = DAGS_DIR + '/cdc_state.json'
with open(state_path) as f:
    state = json.load(f)
print(f'Stored offset: {state.get("last_offset")}  last_run: {state.get("last_run")}')

# Make NO changes to PostgreSQL — next run should detect 0 new events
print('No changes made. Triggering run ...')

run_skip = trigger_dag()
final = poll_run(run_skip, timeout=240)
print(f'\nFinal state: {final}')
show_tasks(run_skip)
````

````python
# Inspect task instances in detail
r = af_get(f'/api/v1/dags/{DAG_ID}/dagRuns/{run_skip}/taskInstances')
tasks = r.json().get('task_instances', [])

print('Expected results:')
print('  health_check      -> success  (connector is RUNNING)')
print('  check_new_events  -> success  (chose the skip_merge branch)')
print('  bronze_cdc        -> skipped  (not chosen by branch)')
print('  silver_cdc        -> skipped  (upstream bronze was skipped)')
print('  skip_merge        -> success  (EmptyOperator, chosen branch)')

print('\nActual results:')
for t in sorted(tasks, key=lambda x: x.get('task_id', '')):
    print(f'  {t["task_id"]:<28} {t.get("state", "none")}')

# Confirm Silver is unchanged
silver_rows = pg('SELECT COUNT(*) FROM silver_customers_af;', fetch=True)[0][0]
print(f'\nsilver_customers_af: {silver_rows} rows (unchanged \u2013 no MERGE ran)')
````

````python
# Now make a change and trigger again — should take the full path
pg("INSERT INTO customers (name, email, country) VALUES ('Eva Svensson', 'eva@example.com', 'Sweden');")
print('Inserted Eva. Waiting for Debezium ...')
time.sleep(5)

run_full = trigger_dag()
final = poll_run(run_full, timeout=240)
print(f'\nFinal state: {final}')
show_tasks(run_full)

print('\nSilver (should include Eva):')
rows = pg('SELECT id, name, country FROM silver_customers_af ORDER BY id;', fetch=True)
for r in rows:
    print(f'  {r}')
````

## Summary

| Concept | Where you saw it | Key point |
|---|---|---|
| `HttpSensor` | `health_check` task | Polls Debezium REST API; fails fast if connector is down |
| Dependency chain | `>>` operator | Enforces execution order; downstream tasks skipped on failure |
| `catchup=False` | DAG definition | Prevents re-running every missed hour on deploy |
| `max_active_runs=1` | DAG definition | No concurrent runs; protects shared state file |
| Retries + backoff | `default_args` | 3 retries with exponential delay; capped at 10 min |
| SLA | `default_args` | Callback fires if task exceeds 30 min |
| Backfill | Part 4 | Reset stored offset \u2192 re-read all Kafka events |
| Idempotency | Bronze + Silver | Same events processed N times = same Silver state |
| Failure & recovery | Part 5 | Sensor retries until connector is healthy again |
| `BranchPythonOperator` | `check_new_events` | Returns task ID at runtime; skips the other path |

**File layout created by this notebook:**
```
dags/
  cdc_pipeline.py      \u2190 DAG code (read by Airflow)
  cdc_state.json       \u2190 stored Kafka offset (written by bronze_cdc task)
  bronze_YYYYMMDD.json \u2190 CDC checkpoint per logical date
```

**Browse the Airflow UI** at [http://localhost:8081](http://localhost:8081) to see the
Graph view, Gantt chart, task logs, and XCom values for each run.
